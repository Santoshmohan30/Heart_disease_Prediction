````python
# Print the results
print("Training set Mean Squared Error (MSE):", train_mse)
print("Training set R-squared (R):", train_r2)
print("Testing set Mean Squared Error (MSE):", test_mse)
print("Testing set R-squared (R):", test_r2)
````
